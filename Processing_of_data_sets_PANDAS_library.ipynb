```python
df.dropna()
```

* Fill empty cells with the given value Unknown

```python
df = df.fillna('Unknown').reset_index(drop=True)
```

* 
Checking for null values

```python
df.isna().sum()
```

##  Data overwrite


```python
df['sex'].unique()
```

* Create a new empty column to correct the value

```python
df['Corect_sex'] = 'empty'
for index, row in df.iterrows():
    if row['sex'] == 'F':
        df.loc[index,'Corect_sex'] = 'Female'
    elif row['sex'] == 'M' :
        df.loc[index,'Corect_sex'] = 'Male'
```

## Rename a column

```python
df = df.rename(columns={"bmd" : "bone_mineral_density"})
df
```

# Discretization of variables


```python
df['age'].describe()
```

```python
bin_labels = ['Baby','Child','Teenager', 'Adult', 'Senior']
cut_bin = [0,1,12,19,60,100]

df['Age_category'] = pd.cut(df['age'], bins=cut_bin, labels=bin_labels)

df
```

# Lambda function

```python
df['height_m'] = df['height_cm'].apply(lambda x: x/100)

df
```

* *Comment* Lambda function is a very helpful tool for mathematical transformations performed on one variable

# Data analysis

```python
df['BMI'] = df['weight_kg']/(df['height_m']*df['height_m'])

df
```

```python
bin_labels = ['starvation','emaciation','underweight', 'correct_value', 'overweight', 'obesity_I', 'obesity_II', 'extreme_obesity']
cut_bin = [0,16,16.99,18.49,24.99,29.99,34.99,39.99,1000]

df['BMI_category'] = pd.cut(df['BMI'], bins=cut_bin, labels=bin_labels)

df
```

```python
df[(df['medication']=='No medication') & (df['Age_category']=='Senior') & (df['BMI_category'] == 'correct_value')]
```

# Data Sorting

```python
df.groupby(['medication']).agg({'Corect_sex': ['value_counts']})
```

* *Comment* 
More women use Anticonvulsant than men abut more men use Glucocorticoids than woman.

```python
df.groupby(['BMI_category']).agg({'age': ['min','mean','max']})
```

* *Comment*  In data base are no obese person with extreme obesity. Thanks to appropriate grouping, it is possible to observe what is the maximum, minimum and average age of people in a given BMI category
